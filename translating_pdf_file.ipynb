## Reading the PDF file

```python
from PyPDF2 import PdfReader


reader = PdfReader("./Google_trans_docs.pdf")
number_of_pages = len(reader.pages)
print(f"Number of pages: {number_of_pages}")
```

## Extracting & Translating the text

```python
from tqdm import tqdm
from googletrans import Translator


translated_text = ""
translator = Translator()

for page in tqdm(reader.pages, total=number_of_pages):
    text = page.extract_text()
    raw_text = repr(text)
    
    translated_text += translator.translate(text=raw_text, src='en', dest='fr').text
    translated_text += "\n"
    
print(translated_text)
```

## Convert RAW text back to Str

```python
translated_text = str(translated_text)
translated_text = translated_text.replace("\\ n", "\n")
translated_text = translated_text.replace("\\ N", "\n")

print(translated_text)
```

## Save the output as a file

```python
with open("translated_text.txt", "w") as file:
    file.write(translated_text)
```
